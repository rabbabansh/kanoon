# DATA RETREIVEL & PRE-PROCESSING

```python
# load data
import os
import shutil
import PyPDF2
import spacy
from pathlib import Path
```

```python
# embeddings
import numpy as np
import pandas as pd
```

```python
#Load spacy English Corpus
!python -m spacy download en_core_web_sm -q
nlp = spacy.load('en_core_web_sm')
```

```python
# Initialize lists for documents, metadatas, and ids
documents = []
metadatas = []
ids = []
```

```python
# Ensure the destination folder /pdfID exists
destination_folder = '../data/out/pdfIDs'
Path(destination_folder).mkdir(parents=True, exist_ok=True)
```

```python
# Function to process each PDF file
def process_pdf(pdf_path, pdfID):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        num_pages = len(reader.pages)
        list_of_pages = [] # List of pages as strings
        
        for i in range(num_pages):
            page_text = reader.pages[i].extract_text()
            list_of_pages.append(page_text)
            
            # Split page into sentences using spacy
            sentences_from_page = [sent.text for sent in nlp(page_text).sents]
            
            for sentenceID, sentence in enumerate(sentences_from_page):
                documents.append(sentence)
                metadatas.append({"pageID": str(i), "pdfID": str(pdfID)})
                ids.append(f"{pdfID}-{i}-{sentenceID}")
    
    return list_of_pages
```

```python
# Iterate through PDF files in /sample_data and process them
source_folder = "../data/source"
pdf_files = [f for f in os.listdir(source_folder) if f.endswith('.pdf') or f.endswith('.PDF')]

for pdfID, pdf_file in enumerate(pdf_files):
    # Process each PDF
    pdf_path = os.path.join(source_folder, pdf_file)
    list_of_pages = process_pdf(pdf_path, pdfID)
    
    # Copy and rename the PDF to /pdfID folder
    destination_path = os.path.join(destination_folder, f"{pdfID}.pdf")
    shutil.copy(pdf_path, destination_path)
```

```python
# Create a dataframe
df = pd.DataFrame({"id": ids, "text": documents, "metadata": metadatas})
df.to_csv('../data/out/pdf_data.csv', index=False)
df.head(50)
```

```python
import json

def fetch_paragraphs_from_pdf(pdfID):
    """
    Fetch all paragraphs from the specified PDF. 
    Assuming the structure of the PDF parsing remains as in 'process_pdf'.
    """
    pdf_path = os.path.join(destination_folder, f"{pdfID}.pdf")
    paragraphs = []
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:  # Ensure there's text on the page
                paragraphs.append(page_text)
    return paragraphs
```

```python
def create_long_prompt(query, results):
    """
    Create a long prompt based on the semantic search query and results, 
    grouping by originating PDF and avoiding duplicate paragraphs.
    """
    documents_by_pdf = {}  # Key: pdfID, Value: list of result documents for that PDF
    
    # Group results by pdfID
    for document, metadata in zip(results['documents'][0], results['metadatas'][0]):
        pdfID = metadata['pdfID']
        if pdfID not in documents_by_pdf:
            documents_by_pdf[pdfID] = {
                "documents": [document],
                "pageIDs": {metadata['pageID']},
            }
        else:
            documents_by_pdf[pdfID]["documents"].append(document)
            documents_by_pdf[pdfID]["pageIDs"].add(metadata['pageID'])
    
    # Convert pageIDs sets to lists for JSON serialization
    for pdfID in documents_by_pdf.keys():
        documents_by_pdf[pdfID]["pageIDs"] = list(documents_by_pdf[pdfID]["pageIDs"])
        paragraphs = fetch_paragraphs_from_pdf(pdfID)
        documents_by_pdf[pdfID]["paragraphs"] = paragraphs
    
    # Construct the final JSON object
    documents_json = json.dumps(documents_by_pdf, indent=4)
    
    # Create the long-prompt including the query and the documents_json
    long_prompt = f"I want you to answer this query '{query}' with all the facts from this data: Don't give too much verbose answer, summarize it, keep it short and it should feel like you're talking to someone rather than a search result formally but not too formally.\n{documents_json}"
    
    return long_prompt
```
